```python

from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("../datasets/mnist_train_small.csv").values
```

```python
X = data[:, 2:].reshape(-1, 28, 28, 1)
y = data[:, 1]
```

```python
generator = ImageDataGenerator(horizontal_flip=True,
                              zoom_range=.2,
                              rotation_range=20,
                              shear_range=.3)
```

```python
gen = generator.flow(X, y)
```

```python
images, values = gen.next()
```

```python
images.shape
```

```python
for image, val in zip(images, values):
    plt.figure()
    plt.title(str(val))
    plt.imshow(image.reshape(28, 28).astype(int), cmap="gray")
```
