```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
```

```python
strange = image.load_img("../lecture_07/strange.jpg", target_size=(200, 200))
```

```python
type(strange)
```

```python
np_img = image.img_to_array(strange)
```

```python
generator = ImageDataGenerator(horizontal_flip=True,
                              zoom_range=.2,
                              rotation_range=20,
                              shear_range=.3)
```

```python
np_batch = np.array([np_img])
```

```python
gen = generator.flow(np_batch)
```

```python
for i in range(5):
    img = gen.next()[0]
    plt.figure()
    plt.imshow(img.astype(int))
```
