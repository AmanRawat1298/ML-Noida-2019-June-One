```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
data = pd.read_csv("../datasets/mnist_train_small.csv").values
```

```python
X = data[:, 2:]
y = data[:, 1]
```

```python
pca = PCA()
```

```python
pca.fit(X)
```

```python
pca.components_.shape
```

```python
X_dash = np.dot(X, pca.components_.T)
```

```python
img = X_dash[0].reshape(28, 28)
```

```python
plt.imshow(img)
```

```python
plt.plot(pca.explained_variance_)
```

```python
np.sum(pca.explained_variance_ > 10000)
```

```python
pca2 = PCA(n_components=2)
```

```python
X_mod = pca2.fit_transform(X)
```

```python
X_select = X_mod[y < 3]
y_select = y[y<3]
```

```python
plt.scatter(X_select[:, 0], X_select[:, 1], c=y_select)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(5)
```

```python
model.fit(X_select, y_select)
```
